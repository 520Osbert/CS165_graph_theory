```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ERDOS_RENYI = "Erdos_Renyi"
BARABASI_ALBERT = "Barabasi_Albert"
sizes = np.array([20, 50, 100, 200, 500, 1000])
colors = ["#24d5c0", "#541706", "#90b4f4", "#79247e", "#42d030", "#f897ac"]

df1 = pd.read_csv("test.csv", sep = ',')
```

```python
Erdos_Renyi_diameter = np.array(df1.loc[df1["model"] == ERDOS_RENYI]["diameter"])
Erdos_Renyi_clustering_coef = np.array(df1.loc[df1["model"] == ERDOS_RENYI]["clustering_coef"])

Barabasi_Albert_diameter = np.array(df1.loc[df1["model"] == BARABASI_ALBERT]["diameter"])
Barabasi_Albert_clustering_coef = np.array(df1.loc[df1["model"] == BARABASI_ALBERT]["clustering_coef"])
```

```python
fig, ax = plt.subplots(figsize = (8, 6))

ax.plot(sizes, Erdos_Renyi_clustering_coef, '.', color = colors[3], label = f'{ERDOS_RENYI}', markersize = 20)
ax.plot(sizes, Erdos_Renyi_clustering_coef, '-',color = colors[3])
ax.plot(sizes, Barabasi_Albert_clustering_coef, '.', color = colors[0], label = f'{BARABASI_ALBERT}', markersize = 20)
ax.plot(sizes, Barabasi_Albert_clustering_coef, '-',color = colors[0])
ax.set_xscale('log', basex=2)

plt.xlabel("log N")
plt.ylabel("C(N) (clustering_coef)")
plt.title("Clustering Coefficients of Two Network Models (Average of 5 samples per N)")
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize = (16, 5))

ax1 = fig.add_subplot(1,2,1)
ax1.plot(sizes, Erdos_Renyi_diameter, '.', color = colors[3], label = f'{ERDOS_RENYI}', markersize = 20)
ax1.plot(sizes, Erdos_Renyi_diameter, '-',color = colors[3])
ax1.set_xscale('log', basex=2)
ax1.set_ylim(2, 6)

plt.xlabel("log N")
plt.ylabel("C(N) (diameter)")
plt.title("Diameters of Erdos-Renyi Network Model (Average of 5 samples per N)\n")

ax2 = fig.add_subplot(1,2,2)
ax2.plot(sizes, Barabasi_Albert_diameter, '.', color = colors[0], label = f'{BARABASI_ALBERT}', markersize = 20)
ax2.plot(sizes, Barabasi_Albert_diameter, '-',color = colors[0])
ax2.set_xscale('log', basex=2)
ax2.set_ylim(2, 6)

plt.xlabel("log N")
plt.ylabel("C(N) (diameter)")
plt.title("Diameters of Barabasi-Albert Network Model (Average of 5 samples per N)\n")

plt.show()
```
